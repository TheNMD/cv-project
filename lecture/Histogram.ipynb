```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# Histogram
img = img = cv.imread('./images/VanGogh.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

counts, bins, bars = plt.hist(img.ravel(), bins=10, range=(0, 256))
plt.show
```

```python
pdf = counts / sum(counts)
cdf = np.cumsum(pdf)

plt.plot(bins[1:], cdf, label="CDF")
plt.legend()
plt.show()
```
